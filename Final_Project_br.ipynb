```python
import pandas as pd
import json
```

```python
moves = pd.read_json('moves.json', orient='records')
moves
```

```python
movement = pd.read_json('movement.json', orient='records')
movement
```

```python
unique = pd.read_json('unique.json')
unique
```

```python
#Export Pandas DataFrames with API Data into Excel
#Method Found at https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_excel.html
with pd.ExcelWriter('output.xlsx') as writer:
    characters.to_excel(writer, sheet_name='characters')
    moves.to_excel(writer, sheet_name='moves')
    movement.to_excel(writer, sheet_name='movement')
    character_attributes.to_excel(writer, sheet_name='character_attributes')
    unique.to_excel(writer, sheet_name='unique')
```

```python
#Create Pivot Table without Aggregation
#Method Found at https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pivot.html
import numpy as np
moves1 = moves.copy()
moves1['HitboxActive'] =moves1['HitboxActive'].astype(str)
moves1['HitboxActive'] = moves1['HitboxActive'].str.split(pat="-")
print(moves1['HitboxActive'])
moves1['HitboxActive'] = moves1['HitboxActive'][0]
moves1['Angle'] = moves1['Angle'].str.split(pat="/")
moves1['Angle'] = moves1['Angle'][0]
moves_pivot = moves1.pivot(index='Owner', columns='Name', values='HitboxActive')
moves_pivot
```

```python
#Create Pivot Table without Aggregation
#Method Found at https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pivot.html
movement1 = movement.copy()
movement1['Value'] = movement['Value'].str.split(pat=" ")

movement_pivot = movement.pivot(index='Owner', columns='Name', values='Value')
movement_pivot
```
