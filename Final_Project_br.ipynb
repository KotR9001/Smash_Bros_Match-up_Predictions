```python
#Create Pivot Table without Aggregation
#Method Found at https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pivot.html
import numpy as np

#Create a Copy of the Table for Operations
moves1 = moves.copy()

#Split the Values for Hitbox Active and Use the First Value
moves1['HitboxActive'] =moves1['HitboxActive'].astype(str)
moves1['HitboxActive'] = moves1['HitboxActive'].str.split(pat="-")
moves1['HitboxActive'] = moves1['HitboxActive'].map(lambda x: x[0])
#Method to Convert Elements Containing Strings Back to Numeric Data Found at https://stackoverflow.com/questions/42719749/pandas-convert-string-to-int
moves1['HitboxActive'] = pd.to_numeric(moves1['HitboxActive'], errors='coerce')
print(moves1['HitboxActive'])

#Create the Pivot Table
hitbox_pivot = moves1.pivot_table(index='Owner', columns='Name', aggfunc=np.mean, values='HitboxActive')
hitbox_pivot = hitbox_pivot[['Airdodge', 'Bair', 'Dair', 'Dash Attack', 'Dash Grab', 'Dsmash', 
                            'Dtilt', 'Fair', 'Fsmash', 'Ftilt', 'Jab 1', 'Jab 2', 'Jab 3',
                            'Nair', 'Spotdodge', 'Standing Grab', 'Uair', 'Usmash', 'Utilt']]
hitbox_pivot
```

```python
#Split the Values for First Actionable Frame and Use the First Value
moves1['FirstActionableFrame'] =moves1['FirstActionableFrame'].astype(str)
moves1['FirstActionableFrame'] = moves1['FirstActionableFrame'].str.split(pat="-")
#Method to Take the First Element from Each Row Found at https://www.w3schools.com/python/python_lambda.asp
moves1['FirstActionableFrame'] = moves1['FirstActionableFrame'].map(lambda x: x[0])
#Method to Convert Elements Containing Strings Back to Numeric Data Found at https://stackoverflow.com/questions/42719749/pandas-convert-string-to-int
moves1['FirstActionableFrame'] = pd.to_numeric(moves1['FirstActionableFrame'], errors='coerce')
print(moves1['FirstActionableFrame'])

#Create the Pivot Table
faf_pivot = moves1.pivot_table(index='Owner', columns='Name', aggfunc=np.mean, values='FirstActionableFrame')
faf_pivot = faf_pivot[['Airdodge', 'Bair', 'Dair', 'Dash Attack', 'Dash Grab', 'Dsmash', 
                            'Dtilt', 'Fair', 'Fsmash', 'Ftilt', 'Jab 1', 'Jab 2', 'Jab 3',
                            'Nair', 'Spotdodge', 'Standing Grab', 'Uair', 'Usmash', 'Utilt']]
faf_pivot
```

```python
#Split the Values for Base Damage and Use the First Value
moves1['BaseDamage'] = moves1['BaseDamage'].astype(str)
moves1['BaseDamage'] = moves1['BaseDamage'].str.split(pat="/")
moves1['BaseDamage'] = moves1['BaseDamage'].map(lambda x: x[0])
#Method to Convert Elements Containing Strings Back to Numeric Data Found at https://stackoverflow.com/questions/42719749/pandas-convert-string-to-int
moves1['BaseDamage'] = pd.to_numeric(moves1['BaseDamage'], errors='coerce')
print(moves1['BaseDamage'])

#Create the Pivot Table
damage_pivot = moves1.pivot_table(index='Owner', columns='Name', aggfunc=np.mean, values='BaseDamage')
damage_pivot = damage_pivot[['Bair', 'Bthrow', 'Dair', 'Dash Attack', 'Dsmash', 'Dthrow',
                            'Dtilt', 'Fair', 'Fsmash', 'Fthrow', 'Ftilt', 'Jab 1', 'Jab 2', 'Jab 3',
                            'Nair', 'Rapid Jab', 'Uair', 'Usmash', 'Uthrow', 'Utilt']]
damage_pivot
```

```python
#Split the Values for Angle and Use the First Value
moves1['Angle'] = moves1['Angle'].astype(str)
moves1['Angle'] = moves1['Angle'].str.split(pat="/")
moves1['Angle'] = moves1['Angle'].map(lambda x: x[0])
#Method to Convert Elements Containing Strings Back to Numeric Data Found at https://stackoverflow.com/questions/42719749/pandas-convert-string-to-int
moves1['Angle'] = pd.to_numeric(moves1['Angle'], errors='coerce')
print(moves1['Angle'])

#Create the Pivot Table
angle_pivot = moves1.pivot_table(index='Owner', columns='Name', aggfunc=np.mean, values='Angle')
angle_pivot = angle_pivot[['Bair', 'Bthrow', 'Dair', 'Dash Attack', 'Dsmash', 'Dthrow',
                            'Dtilt', 'Fair', 'Fsmash', 'Fthrow', 'Ftilt', 'Jab 1', 'Jab 2', 
                            'Nair', 'Rapid Jab', 'Uair', 'Usmash', 'Uthrow', 'Utilt']]
angle_pivot
```

```python
#Split the Values for Base Knockback and Use the First Value
moves1['BaseKnockBackSetKnockback'] = moves1['BaseKnockBackSetKnockback'].astype(str)
moves1['BaseKnockBackSetKnockback'] = moves1['BaseKnockBackSetKnockback'].str.split(pat="/")
moves1['BaseKnockBackSetKnockback'] = moves1['BaseKnockBackSetKnockback'].map(lambda x: x[0])
#Method to Convert Elements Containing Strings Back to Numeric Data Found at https://stackoverflow.com/questions/42719749/pandas-convert-string-to-int
moves1['BaseKnockBackSetKnockback'] = pd.to_numeric(moves1['BaseKnockBackSetKnockback'], errors='coerce')
print(moves1['BaseKnockBackSetKnockback'])

#Create the Pivot Table
knockback_pivot = moves1.pivot_table(index='Owner', columns='Name', aggfunc=np.mean, values='BaseKnockBackSetKnockback')
knockback_pivot = knockback_pivot[['Bair', 'Bthrow', 'Dair', 'Dash Attack', 'Dsmash', 'Dthrow',
                            'Dtilt', 'Fair', 'Fsmash', 'Fthrow', 'Ftilt', 'Jab 1', 'Jab 2', 'Jab 3',
                            'Nair', 'Rapid Jab', 'Uair', 'Usmash', 'Uthrow', 'Utilt']]
knockback_pivot
```

```python
#Create Pivot Table without Aggregation
#Method Found at https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pivot.html

#Create a Copy of the Table for Operations
movement1 = movement.copy()

#Split the Values Using Delimiters
movement1['Value'] = movement1['Value'].str.split(pat=" ")
movement1['Value'] = movement1['Value'].map(lambda x: x[0])

movement_pivot = movement1.pivot(index='Owner', columns='Name', values='Value')
movement_pivot
```

```python
#Create Pivot Table without Aggregation
#Method Found at https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pivot.html

#Create a Copy of the Table for Operations
character_attributes1 = character_attributes.copy()

#Obtain the First Values for Each Element
character_attributes1['Values'] = character_attributes1['Values'].map(lambda x: x[0]['Value'])
print(character_attributes1['Values'])
character_attributes1['Values'] = character_attributes1['Values'].str.split(pat="-")
print(character_attributes1['Values'])
character_attributes1['Values'] = character_attributes1['Values'].map(lambda x: x[0])
print(character_attributes1['Values'])
character_attributes1['Values'] = pd.to_numeric(character_attributes1['Values'], errors='coerce')
print(character_attributes1['Values'])

character_attributes_pivot = character_attributes1.pivot_table(index='Owner', columns='Name', aggfunc=np.mean, values='Values')
character_attributes_pivot = character_attributes_pivot.drop(columns=['Reflectors'])
character_attributes_pivot
```

```python
#Join Hitbox & First Actionable Frame Pivot Tables
joined_hf = hitbox_pivot.join(faf_pivot, on='Owner', how="outer", lsuffix="_hitbox", rsuffix="_faf")
print(len(joined_hf))
joined_hf
```

```python
#Join Damage & Angle Pivot Tables
joined_da = damage_pivot.join(angle_pivot, on='Owner', how="outer", lsuffix="_damage", rsuffix="_angle")
print(len(joined_da))
joined_da
```

```python
#Join the Knockback & Character Attributes Pivot Tables
joined_ka = knockback_pivot.join(character_attributes_pivot, on='Owner', how="outer", lsuffix="_knockback", rsuffix="_character_attributes")
print(len(joined_ka))
joined_ka
```

```python
#Join the Joined Tables
joined_hfda = joined_hf.join(joined_da, on="Owner", how='outer')
joined_inputs = joined_hfda.join(joined_ka, on='Owner', how='outer', lsuffix='_hfda', rsuffix='_ka')
joined_inputs1 = joined_inputs.reset_index()
joined_inputs1
```

```python
#Call in the Character Match-up Chart Data
#Found at https://docs.google.com/spreadsheets/d/1PKEovP2e-slgPFa1ImIoKJBLNsUz3pQMnClUuHdNHsc/edit#gid=954713566
#According to the Information Found at https://www.reddit.com/r/smashbros/comments/8px106/comprehensive_smash_4_matchup_chart_final_update/,
#the Match-up Chart "This is an amalgamation of every notable players matchup chart since 2016, 
#only using data from 2017 and onwards. Averaging this all together, it creates a 55x55 grid of matchup data, 
#tier list based on matchup coverage, and various other stats."

#Also, the Match-up Chart Lack Scores for the Mii Brawler, Mii Gunner, and 'Mii Swordsman'. This is probably because these
#characters were highly customizable.

matchups = pd.read_csv('Smash 4 Comprehensive Matchup Chart - DETAIL.csv')
matchups1 = matchups.dropna(axis=0, how="all")
matchups2 = pd.DataFrame({'Owner':matchups1['Owner'], 'Overall':matchups1['Overall']})
matchups2
```

```python
#Join the Overall Match-up Scores to Create the Final Chart
#joined_final = joined_inputs.set_index('Owner').join(matchups2, on='Owner', how='inner')
#joined_final = pd.concat([joined_inputs1, matchups2], axis=1, join='inner')
joined_final = joined_inputs.merge(matchups2, on='Owner', how='inner').set_index('Owner')
print(f"The number of total values is: {joined_final.count().sum()}")
print(f"The number of null values is: {joined_final.isnull().sum().sum()}")
joined_final
```

```python
#Method to Set Train & Test Data Non-Randomly Found at https://stackoverflow.com/questions/49054538/how-to-split-the-data-set-without-train-test-split
X = joined_final.drop(['Overall'], axis=1)
y = joined_final['Overall']

#NaN Values Needed to Be Removed for Model Fitting, so they Were Dealt With
#Method Found at https://stackoverflow.com/questions/38421624/python-scipy-linear-regression-with-nan-values
#np.isnan checks to see if the Value is NaN and Returns a Boolean of True or False
#Found at https://numpy.org/doc/stable/reference/generated/numpy.isnan.html
#np.logical_not checks to see if the Value is NOT some condition and Returns a Boolean of True or False
#Found at https://numpy.org/doc/stable/reference/generated/numpy.logical_not.html
#The combination of these are in the Range of Elements Returned for X & y
#X=X[X.notna()]
#y=y[y.notna()]

X_train = X.drop(index=['Bayonetta', 'Cloud', 'Corrin', 'Lucas', 'Mewtwo', 'Roy', 'Ryu']) 
X_test = X.loc[['Bayonetta', 'Cloud', 'Corrin', 'Lucas', 'Mewtwo', 'Roy', 'Ryu']]
y_train = y.drop(index=['Bayonetta', 'Cloud', 'Corrin', 'Lucas', 'Mewtwo', 'Roy', 'Ryu']) 
y_test = y.loc[['Bayonetta', 'Cloud', 'Corrin', 'Lucas', 'Mewtwo', 'Roy', 'Ryu']]
```

```python
# Import StandardScaler to prepare data for Fitting to Avoid Biasing the Model to the Testing Data
from sklearn.preprocessing import StandardScaler

#Reshape the Data to Obtain a 2-D Array for the y_scaler
y_train1 = y_train.values.reshape(-1, 1)
y_test1 = y_test.values.reshape(-1, 1)

#Scale to Fit a Gaussian Distribution to the Data
X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train1)

#Transform the Data
X_train_scaled = pd.DataFrame(X_scaler.transform(X_train), columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_scaler.transform(X_test), columns=X_test.columns)
y_train_scaled = pd.DataFrame(y_scaler.transform(y_train1), columns=['Overall'])
y_test_scaled = pd.DataFrame(y_scaler.transform(y_test1), columns=['Overall'])
```

```python
#Create a Model for the Training Data
from sklearn.linear_model import LinearRegression
# explicitly require this experimental feature
from sklearn.experimental import enable_iterative_imputer  # noqa
# now you can import normally from sklearn.impute
from sklearn.impute import IterativeImputer


imputer = IterativeImputer(missing_values=np.nan, add_indicator=False,
                               random_state=0, n_nearest_features=62,
                               sample_posterior=False, verbose=1, initial_strategy='mean',
                          max_iter=10)
imputed = imputer.fit(X_train_scaled)
X_train1 = imputed.transform(X_train_scaled)
X_test1 = imputed.transform(X_test_scaled)


model = LinearRegression()

#joined_final_no_null = joined_final.apply(lambda x: x.fillna(x.median()), axis=0)
#joined_final_no_null
```
