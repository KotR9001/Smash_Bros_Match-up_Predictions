```python
#Create Pivot Table without Aggregation
#Method Found at https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pivot.html


#Create a Copy of the Table for Operations
moves1 = moves_df.copy()

#Split the Values for Hitbox Active and Use the First Value
moves1['HitboxActive'] =moves1['HitboxActive'].astype(str)
moves1['HitboxActive'] = moves1['HitboxActive'].str.split(pat="-")
moves1['HitboxActive'] = moves1['HitboxActive'].map(lambda x: x[0])
#Method to Convert Elements Containing Strings Back to Numeric Data Found at https://stackoverflow.com/questions/42719749/pandas-convert-string-to-int
moves1['HitboxActive'] = pd.to_numeric(moves1['HitboxActive'], errors='coerce')
print(moves1['HitboxActive'])

#Create the Pivot Table
hitbox_pivot = moves1.pivot_table(index='Owner', columns='Name', aggfunc=np.mean, values='HitboxActive')
hitbox_pivot = hitbox_pivot[['Airdodge', 'Bair', 'Dair', 'Dash Attack', 'Dash Grab', 'Dsmash', 
                            'Dtilt', 'Fair', 'Fsmash', 'Ftilt', 'Jab 1', 'Jab 2', 'Jab 3','Light Utilt',
                            'Medium Dtilt', 'Medium Jab 1', 
                            'Nair', 'Spotdodge', 'Standing Grab', 'Uair', 'Usmash', 'Utilt']]
hitbox_pivot
```
